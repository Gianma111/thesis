```python
A_0=np.zeros((N,N))   #prova


for i in range(N):                                          #prova
    for j in range(N):
        if(i == j):
            A_0[i][j] = 0
        else:
            A_0[i][j]=1
#print("A_0=\n",A_0)


A_t=np.zeros((n_time,N,N)) #evoluzione matrice adiacenza

for t in range(n_time):
    A_t[t]=A_0
print("A_t=\n",A_t)
```
