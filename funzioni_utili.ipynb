```python
def coseno_angolo_tra_vettori(x,y,N):  #è corretto definirlo così? nel paper dice che usa la distanza euclidea
    cos = 1                            #NB: i vettori del modello hanno dimensione T (non N,questo è generale)
    X = 0 #modulo di x
    Y = 0 #modulo di y
    scalar_dot = 0 #prodotto scalare
    for i in range(N):
        X = X + x[i]*x[i]
    X = math.sqrt(X)
    print("modulo di X:\n",X)
    for j in range(N):
        Y = Y + y[j]*y[j]
    Y = math.sqrt(Y)
    print("modulo di Y:\n",Y)
    print("modulo X * modulo Y = \n",X*Y)
    for i in range(N):
        scalar_dot = scalar_dot + x[i]*y[i]
    print("prodotto scalare:\n",scalar_dot)
    cos = scalar_dot/(X*Y)
    return cos
```

```python
def estrai_colonna_matrice(mat,N,T,colonna):  #colonna è un intero che mi indica quale colonna della matrice voglio
    
    print("estraggo colonna:",colonna)
    trasposta = mat.T 
    col = trasposta[colonna]
    print("colonna estratta:",col ,"\n tipo:",type(col))
    print("dimensione colonna di matrice:\n",np.shape(col))
    vet_col= col.reshape(-1,1)
    print("trasformo in vettore colonna:\n",vet_col)
    print("dimensioni vettore colonna:\n",np.shape(vet_col))
    
    return vet_col
```

```python
def matrice_di_adiacenza_da_networkx_a_numpy(N,G):
    
    i=0
    j=0
    A = nx.adjacency_matrix(G,nodelist=None,dtype=None,weight =None) #matrice di adiacenza networkx
    A_real = np.zeros((N,N)) #matrice di adiacenza numpy
    
    for i in range(N):
        for j in range(N):
            if(A[[i],[j]] != 0):
                A_real[i][j] =  float(A[[i],[j]])
            else:
                A_real[i][j] =  0
            
    return A_real
```
